```python
import re

def process_question_for_similar_search(question):
    # consecutive question mark
    result = re.compile(r"({})+".format("\?")).sub("?", question)
    print(f"consecutive: {result}")
    
    # alone 
    result = re.compile("\s\?").sub("", result)
    print(f"alone: {result}")
    
    result = str(result).lower().strip()

    # check if question contains multiple sentences
    indexes = [char.start() for char in re.finditer('\?', result)]
    if len(indexes) == 1:
        return result.replace("\?", "").strip()

    return result


def process_question(question):
    """Processes a question for similar search.
    
    Args:
    question: The question to process.
    
    Returns:
    A processed question.
    """
    result = question
    result = re.sub(r"\?+", "?", result) # consecutive mark
    result = re.sub("\s\?", "", result) # standing alone
    result = re.sub(r"\s+", " ", result) # duplicate whitespace
    # result = result.replace("\s+", " ") # duplicate whitespace
    result = result.lower().strip()

    # Check if the question contains multiple sentences.
    if len([char.start() for char in re.finditer('\?', result)]) == 1:
        result = result.replace("?", "").strip()
    
    return result

inp = "what is techday????? ???Who will be the speakers?? ??? ????"
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")

inp = "what is Techday? ????? ????? "
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")

inp = "what is Techday??"
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")

inp = "what is Techday? ? ? ?"
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")

inp = "what is Techday ? ? ? ? "
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")

inp = "what  is Techday ??"
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")

inp = "What  do    you     mean ???? "
res = process_question(inp) 
print(f"INP: {inp}\nRES: {res}\n")
```
